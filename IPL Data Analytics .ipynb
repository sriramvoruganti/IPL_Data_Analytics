```python
#Lets analyze the performance of various Batsmen
```

```python
balls = df.groupby('batsman')['ball'].sum().reset_index()
balls
```

```python
four = df[df['batsman_runs']==4]
four
```

```python
runs_4 = four.groupby('batsman')['batsman_runs'].count().reset_index()
```

```python
runs_4.columns = ['batsman','4s']
```

```python
runs_4 # this table shows the number of fours hit by each batsman
```

```python
#similarly let's calculate the no of sixes
```

```python
six = df[df['batsman_runs']==6]
six
```

```python
runs_6 = six.groupby('batsman')['batsman_runs'].count().reset_index()
```

```python
runs_6.columns = ['batsman','6s']
```

```python
runs_6
```

```python
player = pd.concat([runs,balls.iloc[:,1],runs_4.iloc[:,1]],axis=1)
player
```

```python
player.fillna(0,inplace = True)
```

```python
player
```
